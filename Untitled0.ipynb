Загрузка обучающей выборки

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale = 1./255)
dataBase = datagen.flow_from_directory(
    'InputData/Classes',
    target_size = (80,80),
    batch_size=50,
    class_mode = 'categorical'
)

listOfClasses = dataBase.class_indices
print(listOfClasses)
```

# **Структура нейросети:**
*   Последовательная модель
*   Повторяем два раза с увеличением размера фильтра
 1.   Двумерный свёрточный слой
 2.   Объединяющий слой (Максимальный пулинговый слой)
 3.   Исключающий слой
*   Flatten Для сжатия данных
*   Плотно связанный слой с 512 нейронами
*   Плотно связанный слой с нейронами = количеству классов

Функция активации - relu

Функция потери - Категориальная перекрестная энтропия

Оптимизатор - Адам (так как наиболее высокая производительность)

Метрика - accuracy




# **Свёрточная глубокая нейросеть:**

a.png

```python
import tensorflow as tf

modelNer = tf.keras.models.Sequential([
                                    
    tf.keras.layers.Conv2D(8, (3,3), activation='relu', input_shape = (80, 80, 3)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Conv2D(16, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.2),

    tf.keras.layers.Dense(len(listOfClasses), activation='softmax')
    ])

modelNer.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

print(modelNer.summary())
```

Обучение нейросети

```python
history = modelNer.fit(
    dataBase,
    steps_per_epoch=20,
    epochs = 100)
```

Требуемый фильтр

```python
print('What class or classes do you need?')
print(list(listOfClasses))
print('Examples: "Cars", "Mountains and Pizza"')
filterStr = input()

try:
  filter =[]
  if 'and' in filterStr:  
    for i in filterStr.split(' '):
      if i != 'and':
        filter.append(listOfClasses[i])
  else:
    filter.append(listOfClasses[filterStr])
except Exception:
  raise Exception("Invalid filter value")
```

Название набора для фитрации

```python
import os
print('Which dataset do you want to filter?')
sets=[]
for i_elem in os.listdir(os.path.abspath("/content/InputData/Set")): 
  sets.append(i_elem)
  print(i_elem, end=', ')
print('\nExamples: "Set 1"')
dataSet = input()

if dataSet not in sets:
  raise Exception("Invalid set value")
```

Фильтрация Фотографий

```python
import numpy as np
from keras.preprocessing import image
import numpy
import os
import shutil 

if os.path.exists('/content/Result'):
  shutil.rmtree("/content/Result/")
os.mkdir("/content/Result/")

def get_key(val):
  for key, value in listOfClasses.items():
         if val == value:
             return key

for photo in os.listdir("/content/InputData/Set/" + dataSet):
  if photo.lower().endswith(('.jpg', '.jpeg')):
    path = "/content/InputData/Set/" + dataSet + '/' + photo
    img = image.load_img(path,target_size=(80,80))
    x = image.img_to_array(img)
    x/=255.0
    x = np.expand_dims(x,axis=0)

    classesRes = modelNer.predict(x,batch_size = 50)
    
    if max(classesRes[0]) > 0.7:
      if listOfClasses[get_key(numpy.argmax(classesRes))] in filter:
        shutil.copy(path, "/content/Result/"+ photo) 
```
